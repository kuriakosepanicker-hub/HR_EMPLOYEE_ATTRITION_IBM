# Section 2 - Transformation

We prepared our raw data by first checking for missing values to ensure accuracy. Checking for duplicate values and deleting if any, We then used binary encoding to convert the text answers in our target column into numbers (1 for Yes, 0 for No) so we can perform data math later. Finally, we removed unnecessary columns and saved only our chosen variables into a clean dataset ready for plotting.

```python
#Check for duplicate rows.
print("Duplicate rows found:", df_raw.duplicated().sum())
```

```python
# Remove duplicate rows if any exist
df_cleaned = df_raw.drop_duplicates()
df_cleaned.shape
```

```python
#Check for empty cells (null values)
print("Empty cells found: ", df_cleaned.isnull().sum())




```

```python
#Print unique values.
print(f"Unique values:\n{df_cleaned.nunique()}")
```

```python
df_cleaned.shape
```

```python
#Dropping Columns that are not useful for analysis as they have only one unique value.
columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
df_cleaned = df_cleaned.drop(columns=columns_to_drop)

```

```python
#Checking the columns have been dropped
df_cleaned.shape
```

```python
#creating a copy of the cleaned dataset to be used for analysis.
df_analysis = df_cleaned.copy()
```

## Feature Engineering

Performed binary encoding, we converted the Attrition text column into a new numeric column called "Attrition_Encoded" where every "Yes" was swapped to a 1 and every "No" was swapped to a 0. This simple change unlocked our ability to do math on data and find hidden patterns without changing the true meaning of our data.

* Its clearly depicting that around 1233 employees still working in the organisation,where 237 employees left the organisation.

```python
# Binary encoding: Yes = 1, No = 0
df_analysis['Attrition_Encoded'] = df_analysis["Attrition"].map({"Yes": 1, "No": 0})
df_analysis['Attrition_Encoded'].value_counts()

```

```python
# Now we can take out the Column "Attrition" and keep "Attrition_Encoded" Columninstead.

df_analysis = df_analysis.drop(columns=['Attrition'])
df_analysis.head()
```

# Section 3 - Load

Once the data has been transformed , it will be loaded in to the target analytical environment for further exploration and visualisation.The loading process ensured thatthe cleaned dataset was stored in a structured format.

```python
Col_Final = [
    #original clean columns
    'Age', 'BusinessTravel', 'DailyRate', 'Department', 'DistanceFromHome', 'Education', 'EducationField', 'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager'
#Encoded column
    ,'Attrition_Encoded'
]

print(f"Final columns for analysis:\n{Col_Final}")
df_cleaned = df_analysis[Col_Final].copy()

df_cleaned.to_csv("Dataset/Clean_Dataset/Employee_Attrition_Cleaned.csv", index=False)

df_cleaned.head()
```
